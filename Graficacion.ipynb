```python
fig, axs=plt.subplots(2,2,figsize=(16,10))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0,0].plot(df.loc["Colombia",df_Años])
axs[0,0].set_title("Inmigrantes de Colombia")

axs[0,1].plot(df.loc["Brasil",df_Años])
axs[0,1].set_title("Inmigrantes de Brasil")

axs[1,0].plot(df.loc["Argentina",df_Años])
axs[1,0].set_title("Inmigrantes de Argentina")

axs[1,1].plot(df.loc["Perú",df_Años])
axs[1,1].set_title("Inmigrantes de Perú")

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel("Años")
  ax.set_ylabel("Inmigrantes")

plt.show()
```

#Como los graficos tienen escalas diferentes, para no crear confusiones unificaremos las escalas con ymin y ymax y luego con el ciclo de lazos lo aplicamos a todos los graficos
## Para poner el Titulo en la parte superior de todo el lienzo utilizaremos el llamado .suptitle("Nombre del Titulo"), Para ajustar el tamaño de las fuentes del titulo se utiliza el llamado .fontsize
## Para ajustar el tamaño de los rotulos de los ejes tambien se utiliza el llamado .fontsize en el mimso codigo donde se ponde el nombre
## ymin=0
## ymax=7000
### for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
## Con ax.grid() le colocamos las cuadrillas
##

```python
fig, axs=plt.subplots(2,2,figsize=(16,10))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Inmigracion Latinoamericana a Canada\n Periodo 1980 - 2013", fontsize=18)

axs[0,0].plot(df.loc["Colombia",df_Años])
axs[0,0].set_title("Inmigrantes de Colombia")

axs[0,1].plot(df.loc["Brasil",df_Años])
axs[0,1].set_title("Inmigrantes de Brasil")

axs[1,0].plot(df.loc["Argentina",df_Años])
axs[1,0].set_title("Inmigrantes de Argentina")

axs[1,1].plot(df.loc["Perú",df_Años])
axs[1,1].set_title("Inmigrantes de Perú")

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel("Años",fontsize=14)
  ax.set_ylabel("Inmigrantes",fontsize=14)
  ax.grid()

ymin=0
ymax=7000
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

plt.show()
```
