# Visualizacion de Datos con Python

```python
import pandas as pd
```

```python

df=pd.read_csv("/content/inmigrantes_canada.csv")
```

#Configuracion de un nuevo indice utilizando set_index("nombredecolumna")

```python
df.set_index("Pais",inplace=True)
```

```python
df
```

#Creacion de DF con una lista de todos los años como indice utilizando list(map(str,range(desde,hasta)))

```python
df_Años = list(map(str,range(1980,2014)))
```

```python
df_Años
```

#Creacion de df con los datos del pais que queremos analizar y conjugarlo con el df recien creado. con la funcion .loc ubicamos la linea donde esta la informacion y la extraemos y conjugamos con el otro df

```python
df_Colombia = df.loc["Colombia",df_Años]
```

```python
df_Colombia
```

Mejoramiento del df para mejor manipulacion nombrando las columnas con las que vamos a trabajar

```python
NewList_Colombia = {"Año":df_Colombia.index.tolist(), "Inmigrantes":df_Colombia.values.tolist()}
```

#Una vez creado el diccionario en el punto anterior lo convertimos en Df DatosColombia = pd.DataFrame(nombredeldiccionario)

```python
DatosColombia = pd.DataFrame(NewList_Colombia)
```

```python
DatosColombia
```

```python
import matplotlib.pyplot as plt
```

#Graficacion de los datos con el llamado .plot(nombredeldf["nombredecolumnapara x"],nombredeldf["nombredecolumnapara y"])

```python
plt.plot(DatosColombia["Año"],DatosColombia["Inmigrantes"])
```

#Mejoramiento de la visualizacion del grafico
## Parametros de los intervalos se mejoran con .xticks(["1980", "1985" , "1990", "1995", "2000", "2005", "2010", "2015"])
#Para quitar este rotulo predeterminado [<matplotlib.lines.Line2D at 0x7e792154ade0>] lo hacemos con .show
#Para agrandar el grafico lo hacemos con .figure(figsize=(Alto,Ancho)
##NOTA: Se debe configurar el tamaño antes

```python
plt.figure(figsize=(8,5))
plt.plot(DatosColombia["Año"],DatosColombia["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010","2015"])
plt.show()
```

#Agreguemos el titulo al grafico con el llamado .title("NombreTitulo")
#Agreguemos rotulos a los ejes con el llamado .xlabel("Nombre del Eje") ylabel("Nombre del Eje")

```python
plt.figure(figsize=(8,5))
plt.plot(DatosColombia["Año"],DatosColombia["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010","2015"])
plt.title("Inmigrantes de Colombia")
plt.xlabel("Años")
plt.ylabel("Inmigrantes")
plt.show()
```

#Para trabajar es mejor mantener los graficos en funciones, de tal manera de llamarla en cualquier momento y que sea dinamica.
Para ello se utiliza el llamado fig,ax=plt.subplots(figsize=(8,4))
ax.plot(DatosColombia["Año"],DatosColombia["Inmigrantes"])

```python
import matplotlib.ticker as ticker

fig,ax=plt.subplots(figsize=(8,4))
ax.plot(DatosColombia["Año"],DatosColombia["Inmigrantes"])
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.set_title("Inmigrantes de Colombianos Hacia Canada\n1980 - 2013")
ax.set_xlabel("Años")
ax.set_ylabel("Inmigrantes")

plt.show()
```
